```python
import pandas as pd
import numpy as np
import netCDF4
from netCDF4 import Dataset
import matplotlib.pyplot as plt
import os
```

```python
dir_ozone = '../data/Ozone/'
dir_figure = '../figures/ozone/'

filename_prev = 'OMI-Aura_L3-OMDOAO3e_'
filenames = [filename for filename in os.listdir(dir_ozone) if filename_prev in filename]
filenames = sorted(filenames)
```

```python
for filename in filenames:
    date = filenames[0][21:30]
    year = date[:4]
    month = date[-4:-2]
    day = date[-2:]
    f = Dataset(dir_ozone + filename)

    lons = f.variables['lon'][:].data
    lons[lons > 180] -= 360
    lats = f.variables['lat'][:].data
    lons, lats = np.meshgrid(lons, lats)
    values = f.variables['ColumnAmountO3'][:].data
    unit = f.variables['ColumnAmountO3'].Units

    iwant = (values > 0)

    plt.rcParams["figure.figsize"] = (12, 6) # (w, h)
    plt.rcParams.update({'font.size': 20})

    plt.scatter(lons[iwant], lats[iwant], c = values[iwant], cmap = 'jet', s = 1)
    plt.colorbar()
    plt.title('Column Amount $O_3$ / ' + unit + ': ' + year +'-'+ month + '-' + day)
    plt.xlabel('longitude')
    plt.ylabel('latitude')
    plt.xticks(ticks = np.array(range(-180,181,60)), labels = np.array(range(-180,181,60)))
    plt.yticks(ticks = np.array(range(-90,90,30)), labels = np.array(range(-90,90,30)))
    plt.savefig(dir_figure + year +'_'+ month + '_' + day + '.png')
    plt.close()
    break
```
