## Read Data

```python
days = []
for i in range(1,31+1):
    day = str(i)
    if len(day)<2:
        day = '0'+day
    days.append(day)

days
```

```python
# GPS data
node_id = '001e0610c2e9'
dir_in = '../Minolta/'+ node_id + '/'
sensor_name = 'GPGGA'

df_gps = []
for year in ['2020']:#years:
    for month in ['01','02']:#months:
        for day in days:
            if len(day) == 1:
                day = '0' + day
            dirname = dir_in+year+'/'+month+'/'+day+'/'
            if not os.path.isdir(dirname):
                continue
            print(dirname)
            
            filename = dirname+'MINTS_'+node_id+'_'+sensor_name+'_'+year+'_'+month+'_'+day+'.csv'
            if not os.path.isfile(filename):
                continue
            df1 = pd.read_csv(filename, parse_dates=True, index_col = 'dateTime')
            df1.index.name = 'UTC'
            df1 = df1[['latitude', 'longitude', 'altitude']]
            # merge into df
            if len(df_gps)==0:
                df_gps = df1
            else:
                df_gps = pd.concat([df_gps, df1])
```

```python
display(df_gps)
df_gps['latitude'] = (df_gps['latitude'] // 100) + (df_gps['latitude'] % 100)/60
df_gps['longitude'] = (df_gps['longitude'] // 100) + (df_gps['longitude'] % 100)/60
display(df_gps)
```

```python
df_gps[['latitude']].median()
```

```python
lat_median = float(df_gps[['latitude']].median()) # 32.992192
lat_delta = 0.0001

plt.rcParams["figure.figsize"] = (30,8)
df_gps[['latitude']].plot()
plt.xlim(datetime.date(2020, 1, 6),datetime.date(2020, 3, 1))
plt.ylim(lat_median-lat_delta, lat_median+lat_delta*2)
```

```python
long_median = float(df_gps[['longitude']].median()) # 32.992192
long_delta = 0.0001

plt.rcParams["figure.figsize"] = (30,8)
df_gps[['longitude']].plot()
plt.xlim(datetime.date(2020, 1, 6),datetime.date(2020, 3, 1))
plt.ylim(long_median-long_delta, long_median+long_delta)
```
