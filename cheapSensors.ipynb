```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates

dir_out = '../figures/'

year = 2019
month = 12
day = 25

hour_start_local = 6
hour_end_local = 18

hour_jetlag = 6
jetlag = datetime.timedelta(hours=hour_jetlag)

df_TSL = pd.read_csv("../lightsensors/001e06305a6b/%02d/%02d/%02d/MINTS_001e06305a6b_TSL2591_%02d_%02d_%02d.csv" % (year, month, day, year, month, day))
df_TSL = df_TSL.set_index('dateTime')

df_TSL.index = pd.to_datetime(df_TSL.index)

display(df_TSL.head())


datetime_start = (datetime.datetime(year, month, day, hour_start_local, 0, 0) + jetlag)
datetime_end   = (datetime.datetime(year, month, day, hour_end_local,   0, 0) + jetlag)
iwant = df_TSL.index > datetime_start
iwant*= df_TSL.index < datetime_end
df_iwant = df_TSL[iwant].copy()
#if len(df_iwant)==0:
#    continue


df_TSL.index = pd.to_datetime(df_TSL.index)
df_TSL.index = df_TSL.index - jetlag

x = df_TSL.index# local time #.hour.values[:]

for var in df_TSL.columns:
    print(var)
    y = df_TSL[var].values
            
    fig, ax = plt.subplots(constrained_layout=True, figsize=(20, 10))
    h = ax.plot(x,y)
            
    ax.set_title( var +': %02d/%02d/%02d' % (year,month,day))
    ax.set_xlabel('Time')
    ax.set_ylabel('Count')
    
    locator = mdates.HourLocator(interval = 1)
    h_fmt = mdates.DateFormatter('%H:%M')
    ax.xaxis.set_major_locator(locator)
    ax.xaxis.set_major_formatter(h_fmt)
    fig.autofmt_xdate()
    fig.savefig(dir_out+'TSL/'+ var + '_%02d_%02d_%02d.png' % (year, month, day))
    plt.close()
```
